## 2.1 Series

```python
# Creatin a series from a numpy array
temperature_values = np.array([15.2, 16.5, 14.8, 17.0, 16.3, 15.6, 14.9])

# Creating the Series
daily_max_temps = pd.Series(temperature_values)

daily_max_temps

# Check index and values attributes
```

```python
# create a date-based index
dates = pd.date_range(start='2025-01-16', periods=7, freq='D')
daily_max_temps = pd.Series(data=temperature_values, index=dates)
daily_max_temps
```

```python
# A series from a dictionary
# Dictionary containing daily rainfall measurements
rainfall_data = {
    '2025-01-16': 5.2,
    '2025-01-17': 0.0,
    '2025-01-18': 12.4,
    '2025-01-19': 7.8,
    '2025-01-20': 0.0,
    '2025-01-21': 3.6,
    '2025-01-22': 0.0
}

# Creating the Series from the dictionary
daily_rainfall = pd.Series(data=rainfall_data, name='Daily Rainfall (mm)')

daily_rainfall
```

## 2.2 DataFrame

```python
# Creating DataFrame from Dictionary of arrays
# Daily Max and Min Temperatures
temperature_data = {
    'Max Temperature (C)': [13, 14, 17.1, 17.8, 15.3, 13.7, 16.5],
    'Min Temperature (C)': [6.8, 7.3, 8.3, 8.0, 7.5, 7.4, 8.9]
}

# Creating the DataFrame
temperature_df = pd.DataFrame(data=temperature_data, index=dates)

temperature_df
```

```python
# Creating DataFrame from Dictionary of Series

max_temperature = pd.Series(data=[13, 14, 17.1, 17.8, 15.3, 13.7, 16.5], index=dates)

min_temperature = pd.Series(data=[6.8, 7.3, 8.3, 8.0, 7.5, 7.4, 8.9], index=dates)

# Combining the Series into a DataFrame with explicit column names
temperature_df = pd.DataFrame({'Max Temperature (C)': max_temperature,  'Min Temperature (C)': min_temperature})

# Displaying the DataFrame
temperature_df
```

# 4. Data Preparation
## 4.1 Handling Missing Data

```python
# replace all NaNs by -9999
housing_9999 = housing_data.fillna(-9999)
housing_9999[housing_9999["total_bedrooms"] == -9999]
```

```python
# Fill missing values in 'total_bedrooms' with the mean of the column
housing_data['total_bedrooms'].fillna(housing_data['total_bedrooms'].mean(), inplace=True)
# Check for missing values
print(housing_data.isnull().sum())
```

## 4.2 Feature Engineering
- Creating New Features

```python
# Create a new column for population density (average population per household)
# Creating population density allows exploring the relationship between population patterns and house values in a more meaningful way.
housing_data['population_density'] = housing_data['population'] / housing_data['households']

# Display the first few rows to confirm the preprocessing
housing_data.head()
```

# Question 1:
## What are the key factors that affect the median house value in a particular area?
### Solution
* Calculate the correlation between median house value and other numerical features.
* Shows how strongly each variable correlates with the median house value

```python
# Drop non-numeric columns before calculating the correlation matrix
housing_data_numeric = housing_data.drop(columns=['ocean_proximity'])

# Calculate the correlation matrix for numeric columns
correlation_matrix = housing_data_numeric.corr()

# Extract the correlation of each variable with the median_house_value
correlation_with_house_value = correlation_matrix["median_house_value"].sort_values(ascending=False)

# Display the top correlations
correlation_with_house_value
```

## What does this mean?

* **Median income** has the highest positive correlation (0.688); higher-income areas tend to have higher house values.
* **Total rooms** shows a weak positive correlation (0.134); areas with more rooms may have slightly higher house values
* **Latitude** has a weak negative correlation (-0.144); house values tend to decrease as you move further from certain geographic regions (likely away from the coast).
* Other variables like total bedrooms, population, and longitude have very weak or negligible correlations.

# One-Hot Encoding (optional)
- We can include ocean_proximity in the analysis by using one-hot encoding to convert the categorical values into numerical

```python
# Convert categorical column 'ocean_proximity' to numerical values using one-hot encoding
housing_data_encoded = pd.get_dummies(housing_data, columns=['ocean_proximity'], drop_first=True)

# Calculate the correlation matrix with encoded data
correlation_matrix_encoded = housing_data_encoded.corr()

# Extract the correlation of each variable with the median_house_value
correlation_with_house_value_encoded = correlation_matrix_encoded["median_house_value"].sort_values(ascending=False)

# Display the top correlations
correlation_with_house_value_encoded
```

# Question 2
# How does proximity to the ocean impact house prices?
**Group** the dataset by the **ocean proximity** variable and analyze the **average house value** for each category

```python
# Group by ocean_proximity and calculate the average median_house_value for each category
average_house_value_by_ocean_proximity = housing_data.groupby('ocean_proximity')['median_house_value'].mean()

# Display the results
average_house_value_by_ocean_proximity.sort_values(ascending=False)
```

## What do these values tell us?
Proximity to the ocean tends to increase house values
